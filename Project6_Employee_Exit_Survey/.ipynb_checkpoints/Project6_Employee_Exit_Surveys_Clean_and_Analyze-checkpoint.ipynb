## Job Dissatisfaction

Both dataframes contain multiple boolean statements pertaining to job dissatisfaction. We will use this in combination with the "any()" method to create a new column called "dissatisfied". If any factor leads to dissatisfaction then the corresponding value in "dissatisfied" will be set to true. 


Lets view the dissatisfaction frequency inside the 'Contributing Factors. Dissatisfaction' and 'Contributing Factors. Job Dissatisfaction' columns of the tefe DataFrame

```python
tafe_resignations["Contributing Factors. Dissatisfaction"].value_counts(dropna=False)
```

```python
tafe_resignations["Contributing Factors. Job Dissatisfaction"].value_counts(dropna=False)
```

Clean both of the columns by replacing their values to True, False or NaN

```python
def update_vals(val):
    if pd.isnull(val):
        return np.nan
    elif val == "-":
        return False
    else:
        return True

tafe_resignations[["Contributing Factors. Dissatisfaction","Contributing Factors. Job Dissatisfaction"]]= tafe_resignations[["Contributing Factors. Dissatisfaction","Contributing Factors. Job Dissatisfaction"]].applymap(update_vals)
```

```python
tafe_resignations.head()
```

```python
tafe_resignations["dissatisfied"] = tafe_resignations[["Contributing Factors. Dissatisfaction","Contributing Factors. Job Dissatisfaction"]].any(axis = 1,skipna=False)
tafe_resignations["dissatisfied"].head(10)
```

```python
dete_resignations["dissatisfied"] = dete_resignations[["job_dissatisfaction","dissatisfaction_with_the_department","physical_work_environment","lack_of_recognition","lack_of_job_security","work_location","employment_conditions","work_life_balance","workload"]].any(axis=1,skipna=False)
dete_resignations["dissatisfied"].head(10)
```

Copy both dataframes, we will be using these copies to merge with each other.

```python
dete_resignations_up = dete_resignations.copy()
tafe_resignations_up = tafe_resignations.copy()
```

## Merging TAFE and DETE Datasets

In this section we will be merging the TAFE and DETE Datasets via the institute_service column.

```python
dete_resignations_up["institute"] = "TEFE"
tafe_resignations_up["institute"] = "DETE"
```

```python
combined = pd.concat([dete_resignations_up,tafe_resignations_up])

combined.shape
#combined.head()
```

Next, we will drop any any columns with less that 500 non-null values using the "thresh" parameter

```python
combined.isnull().sum()
```

```python
combined_updated = combined.dropna(axis = 1,thresh=500)
```

```python
combined_updated.isnull().sum()
```

In the next section we will be cleaning up the "institute_service" column from our combined_updated dataset. Using this article: https://www.businesswire.com/news/home/20171108006002/en/Age-Number-Engage-Employees-Career-Stage ,
we will make the argument that there is a correlation between an employee's needs and career stage instead of age. The institute_service column will therefore be re-categorized as follows on a new row:

- New: Less than 3 years at a company
- Experienced: 3-6 years at a company
- Established: 7-10 years at a company
- Veteran: 11 or more years at a company

```python


combined_updated["institute_service"].value_counts()
```

Create a regex pattern that will extract numbers from strings and then apply it to the "institute_service" column

```python
pattern = '(\d+)'

combined_up = combined_updated["institute_service"].astype(str).str.extract(pattern).astype(float)
combined_up[0].value_counts()
```

The next step involves creating a function that maps each year value into a career stage described above.

```python
def map_stage(val):
    
    if pd.isnull(val):
        return np.nan
    elif val < 3 :
        val =  "New"
        return val
    elif val>= 3 and val <= 6:
        val =  "Experienced"
        return val    
    elif val>= 7 and val <= 10:
        val =  "Established"
        return val   
    elif val >=11:
        val =  "Veteran"
        return val      

career = combined_up[0].apply(map_stage)
```

```python
combi_fin= combined_updated.copy()

combi_fin["service_cat"]=career
```

```python
combi_fin
```

## Analysis

We can now analyze and see how many employees resigned due to dissatisfaction

```python
combi_fin["dissatisfied"].value_counts(dropna=False)
```

There were more lmost double of dissatisfied employees compared to satisifed employees. We will handle the missing values by replacing them with the average, which in this case is False.

```python
combi_fin.fillna(False).head()
```

```python
# Convert the dissatisfied column into a boolean before running the pivot_table() function
combi_fin["dissatisfied"]=combi_fin["dissatisfied"].astype(bool)
combi_fin["dissatisfied"]
```

```python
service_diss = combi_fin.pivot_table(values = "dissatisfied", index="service_cat" )
service_diss
```

We can tell that the majority of employees who terminated their employment were either established or veneran employees. Newer employees tended to have other reasons for terminating their employment, this should warrant further investigation based on the questionnaire.
