## Cleaning the Datasets

We replaced the "Not Stated" values in the DETE survey with NaN value by re-reading DETE csv above using the na_values parameter. 
The next few cells will show how we dropped unnecessary columns from our dataset. 
- Drop Columns 28-49 in DETE survey
- Drop columns 17-66 in TAFE survey

```python
dete_survey_updated = dete_survey.drop(dete_survey.columns[28:49],axis = 1)
```

```python
tafe_survey_updated = tafe_survey.drop(tafe_survey.columns[17:66],axis = 1)
```

Before combining both of the exit survey datasets we will standardize the column names by making them lower case and replacing spaces with an underscore.

```python
dete_survey_updated.columns = dete_survey_updated.columns.str.lower().str.strip().str.replace(" ","_")
```

```python
dete_survey_updated.columns
```

Next, we standardize the column names by using the `rename()` method

```python
col_name_dict= {'Record ID': 'id','CESSATION YEAR': 'cease_date','Reason for ceasing employment': 'separationtype',
               'Gender. What is your Gender?': 'gender','CurrentAge. Current Age': 'age','Employment Type. Employment Type': 'employment_status',
               'Classification. Classification': 'position',
               'LengthofServiceOverall. Overall Length of Service at Institute (in years)': 'institute_service',
                'LengthofServiceCurrent. Length of Service at current workplace (in years)': 'role_service'}
```

```python
tafe_survey_updated.rename(columns=col_name_dict, inplace=True)
```
