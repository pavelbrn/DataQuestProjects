### Cleaning the Survey Data

Consider the first two columns: 
- Have you seen any of the 6 films in the Star Wars franchise?
- Do you consider yourself to be a fan of the Star Wars film franchise?

These contain "Yes", "No" and NaN values. We will replace those values with True, False and keep NaN.

```python
yes_no_dict = {"Yes":True,"No":False}
```

```python
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"]=star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no_dict)
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"]= star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no_dict)

star_wars.head()
```

Lets examine the fourth column: 
- Which of the following Star Wars films have you seen? Please select all that apply.

This columns and the next 5 columns after it all contain movies that the user has seen. For readibility purposes we will rename the column mentioned above to "seen_1", "seen_2"..."seen_n" and change their values to True or False instead of naming the movie.




```python


new_names={"Which of the following Star Wars films have you seen? Please select all that apply.":"seen_1",
          "Unnamed: 4": "seen_2", "Unnamed: 5":"seen_3",
          "Unnamed: 6":"seen_4","Unnamed: 7":"seen_5",
          "Unnamed: 8":"seen_6"}

star_wars.rename(columns=new_names,inplace = True)




col_names = star_wars.columns[3:9]

val_dict={"Star Wars: Episode I  The Phantom Menace":True,
         "Star Wars: Episode II  Attack of the Clones":True,
         "Star Wars: Episode III  Revenge of the Sith":True,
         "Star Wars: Episode IV  A New Hope":True,
         "Star Wars: Episode V The Empire Strikes Back":True,
         "Star Wars: Episode VI Return of the Jedi":True}

#print(col_names)

for col in col_names:
    star_wars[col]= star_wars[col].map(val_dict)
```

```python
star_wars.head(5)
```

Columns 9-15 ask the user to rate their favorite Star Wars movie on a scale of 1-6. The current format of these values is on Object form, we will convert these to a nueric format. 


```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
rename_dict = {"Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.":"ranking_1",
               "Unnamed: 10":"ranking_2",
              "Unnamed: 11":"ranking_3",
              "Unnamed: 12":"ranking_4",
              "Unnamed: 13":"ranking_5",
              "Unnamed: 14":"ranking_6"}
```

```python
star_wars.rename(columns=rename_dict,inplace = True)
```

```python
star_wars.head()
```

### Ranking Most Popular Movies

```python
ranking_bar = star_wars.iloc[:,9:15].mean()
print(ranking_bar)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patheffects as PathEffects
%matplotlib inline
```

```python
ranking_bar.plot.bar(color= "#F9D50E",edgecolor= "black",linewidth= 2)
txt = plt.text(-0.5,5,'Star Wars Movie Rankings', size=25, color='#F9D50E')
txt.set_path_effects([PathEffects.withStroke(linewidth=3, foreground='black')])
plt.draw()

plt.xlabel("Movies Part 1-6")
plt.ylabel("Average Score")
plt.show()

```

The majority of the users selected is "Star Wars: Episode V The Empire Strikes Back", just as hypothesized by the editors of FiveThirtyEight.

In the next part we will sum the boolean values from the "seen" columns and visualize which movies were seen the most.

```python
most_seen= star_wars.iloc[:,3:9].sum(axis =0)

most_seen.head(10)
```

```python
most_seen.plot.bar(color= "#F9D50E",edgecolor= "black",linewidth= 2)
txt = plt.text(-1,850,'Most Popular Star Wars Movies', size=25, color='#F9D50E')
txt.set_path_effects([PathEffects.withStroke(linewidth=3, foreground='black')])
plt.draw()

plt.xlabel("Seen Movies Part 1-6")
plt.ylabel("Number of Views")



```

The majority of surveyed user have also seen the The Empire Strikes Back, further confirming the hypothesis that it is the most popular movie in america.

### Movie Popularity: Women vs. Men

In this section we will split the main dataframe into males and females. We will use this split to answer the following questions:
- What is the most vied film amongs women and men?
- Do men and women rank Star wars movies differently?

```python
star_women = star_wars[star_wars["Gender"] == "Female"]
star_men = star_wars[star_wars["Gender"] == "Male"]
```

In this section we will analyze the female view count and popularity of Star Wars movies.

```python
women_r = star_women.iloc[:,9:15].mean()
```

```python
women_r.plot.bar()
plt.show()
```

```python
women_p = star_women.iloc[:,3:9].sum()
women_p
```

```python
women_p.plot.bar()
plt.show()
```

```python
men_p.plot.bar()
plt.show()
```
