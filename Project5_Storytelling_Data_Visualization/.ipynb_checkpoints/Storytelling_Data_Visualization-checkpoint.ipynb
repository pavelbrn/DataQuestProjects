# Storytelling Data Visualization

The following currency exchange dataset will be used to show storytelling data visualization in Matplotlib, more epecifically, we will focus on the exchange rate between the euro and the American dollar.

This dataset was put togeather by Daria Chemkaeva on Kaggle and is available on:https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020

## Importing and Exploring the Dataset

```python
import pandas as pd
exchange_rates= pd.read_csv("euro-daily-hist_1999_2020.csv")

import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.style as style
```

```python
exchange_rates.head()
```

```python
exchange_rates.info()
exchange_rates.isnull().sum()
```

There are 40 columns representing individual countries, 17 columns contiain null values that are also not in float or integer form but are representd as objects. 

## Cleaning the Data

The dataset will be cleaned in the following way:
- Rename [US dollar ] and Period\Unit to US_dollar and Time
- Change Time to datetime data type
- Sort time column in ascending order
- Reset the index


```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
exchange_rates.head(20)
```

In the cell below we isolate only the Time and US_dollar columns

```python
euro_to_dollar = exchange_rates.loc[:,["Time","US_dollar"]]
euro_to_dollar.info()
```

```python
euro_to_dollar["US_dollar"].value_counts()
```

After exploring the frequency of the US_dollar column we can see that there is a non-null value that was hidden from us because it was allocated to a "-" symbol.
We will remove this from our dataset and replace it will a null value.

```python
euro_to_dollar= euro_to_dollar[euro_to_dollar["US_dollar"] != "-"]
euro_to_dollar.head(20)

# Check if the "-" values were removed, by re running the vaue_counts method
euro_to_dollar["US_dollar"].value_counts()
```

Next, we convert all values inside the US_dollar column to float

```python
euro_to_dollar["US_dollar"]= euro_to_dollar["US_dollar"].astype(float)
plt.plot(euro_to_dollar["Time"], euro_to_dollar["US_dollar"])
plt.show()
```

```python
euro_to_dollar.info()
euro_to_dollar
```

## Data Visualization

In this section we will show how the euro-dollar rate has changed during the 2007-2008 financial crisis using a comparison plot for the years 2006-2009.

We will visualize our dataset using the Matplotlib library. First, we will create a rolling mean column to smooth out our line graph.

```python
euro_to_dollar["rolling_mean"]= euro_to_dollar["US_dollar"].rolling(30).mean()
euro_to_dollar

plt.plot(euro_to_dollar["Time"], euro_to_dollar["rolling_mean"])
plt.show()
```

Plot generation:
- Extract a series from 2006 to 2009
-- update the "Time" column to show just years
-- extract all rows with years between 2006 and 2009

- Plot Exchange Rate versus Time
- Highlight the years between 2007-2008
- turn the line section between 2007 and 2009 to have a yellow color

```python
financial_06_09=euro_to_dollar.copy()[(euro_to_dollar["Time"].dt.year>=2006) & (euro_to_dollar["Time"].dt.year<=2009)]

financial_07_08=euro_to_dollar.copy()[(euro_to_dollar["Time"].dt.year>=2007) & (euro_to_dollar["Time"].dt.year<=2008)]
```

```python
style.use('ggplot')

# Set the figure size:
fig, ax =plt.subplots(figsize=(10,5))


# Clear all x and y ticklabels
ax.set_xticklabels([])
ax.set_yticklabels([])

# To fix the ValueError:
# run print(ax.get_xticks())
# check the first and second tick, subtract second tick from first
# to get the difference and use this as the += value to increment x.

# Set new ticklabels base on x tick values
x=732312
for i in ['2006', '2007', '2008', '2009', '2010']:
    ax.text(x, 1.13, i, alpha=0.5, fontsize=11)
    x+=362

y = 1.15
for i in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(732312, y, i, alpha=0.5, fontsize=11)
    y += 0.135

# Plot both line graphs
ax.plot(financial_06_09["Time"], financial_06_09["rolling_mean"])

ax.plot(financial_07_08["Time"], financial_07_08["rolling_mean"])

# Hightlight the peak of the financial crisis
ax.axvspan(xmin=733100, xmax=733298, ymin=0.0,
           alpha=0.3, color='grey')

# Add labels
ax.text(x=732412, y=1.61,s="Euro-USD Exchange Rate During the 2007-2008 Financial Crisis",size=15,weight="bold")


plt.show()
```
